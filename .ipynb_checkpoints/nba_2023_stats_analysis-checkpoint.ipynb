```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Source dataset: https://www.kaggle.com/datasets/amirhosseinmirzaie/nba-players-stats2023-season

<h1>Load Dataset</h1>

```python
df = pd.read_csv('./2023_nba_player_stats.csv')

df.head()
```

```python
# Preview all columns in the dataset
df.columns.values
```

```python
# Calculate average triple doubles and points-per-game
df['AVG_TD3'] = df['TD3']/df['GP']
df['PPG'] = df['PTS']/df['GP']
df
```

```python
# Min-Max Scaling for PTS and PPG
df['Scaled_PTS'] = (df['PTS'] - df['PTS'].min()) / (df['PTS'].max() - df['PTS'].min())
df['Scaled_PPG'] = (df['PPG'] - df['PPG'].min()) / (df['PPG'].max() - df['PPG'].min())

# Plotting Scaled PTS and PPG side by side
plt.figure(figsize=(12, 6))
plt.scatter(df['PName'].head(10), df['Scaled_PTS'].head(10), color='blue', label='Scaled PTS', alpha=0.7)
plt.scatter(df['PName'].head(10), df['Scaled_PPG'].head(10), color='red', label='Scaled PPG', alpha=0.7)
plt.xticks(rotation=90)
plt.legend()
plt.title('Comparison of Scaled PTS and Scaled PPG')
plt.tight_layout()
plt.show()
```

```python
# Calculate top 5 players based on age
df_top_5_age = df.sort_values(by=['Age'],ascending=False).head(5)
df_top_5_age.plot(x="PName", y=["Age"], kind="bar")
```

```python
# Determine top 5 for total points
df_top_5_points = df.sort_values(by=['PTS'],ascending=False).head(5)
df_top_5_points.plot(x="PName", y=["PTS"], kind="bar")
```

```python
# Determine top 5 for average triple doubles
df_top_5_avg_td3 = df.sort_values(by=['AVG_TD3'],ascending=False).head(5)
df_top_5_avg_td3.plot(x="PName", y=["AVG_TD3"], kind="bar")
```

```python
# Selected pairs for correlation analysis
pairs = [
    ('PTS', 'FG%'),
    ('3PM', '3P%'),
    ('AST', 'TOV'),
    ('STL', 'BLK'),
    ('DREB', 'BLK'),
    ('REB', 'PF'),
    ('DD2', 'TD3'),
    ('PTS', '+/-'),
    ('AST', '+/-')
]

# Calculate correlation for selected pairs
correlation_values = [(pair[0], pair[1], df[pair[0]].corr(df[pair[1]])) for pair in pairs]

# Plotting
plt.figure(figsize=(10, 8))
sns.barplot(x=[f"{pair[0]} vs {pair[1]}" for pair in pairs], y=[correlation for _, _, correlation in correlation_values])
plt.title('Correlation Analysis for Selected Stat Pairs')
plt.ylabel('Correlation Coefficient')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
